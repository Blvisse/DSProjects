```python
""""
 we create a function that does the following
 1. calculates the quantile statistcis of all the columns
 2.calculates the descripritve stats of the columns
 3.creates a histogram of the columns 

"""

#we can creat a dictionary that stores each of the columns and their respective values hence creating a dataframe with the info


def quantileStats(rows):
  stats={}
  for row in rows.columns:
    stats[row]=rows[row].describe()
    statsPD=pd.DataFrame(stats)
  return statsPD
```

```python
quantileStats(num)
```

```python
def DescriptiveStats(rows):
  dStats={}
  for row in rows.columns:
    dStats[row]={'skew':stats.skew(rows[row]),'kurtosis':stats.kurtosis(rows[row])}
    dStatsPD=pd.DataFrame(dStats)

  return dStatsPD
```

```python
stats.skew(train.FQ1)
```

```python
descriptiveDF=DescriptiveStats(num)
descriptiveDF=descriptiveDF.T
descriptiveDF
```

* **If the skewness is between -0.5 and 0.5, the data are fairly symmetrical**
* **If the skewness is between -1 and — 0.5 or between 0.5 and 1, the data are moderately skewed**
* **If the skewness is less than -1 or greater than 1, the data are highly skewed**

```python
#lets get the higly skewed columns

descriptiveDF[(descriptiveDF['skew']>1)| (descriptiveDF['skew']<-1) ]
```

* If the absolute value of skew < 0.5 then very symmetric.
* If the absolute value of skew is in between 0.5 and 1 then slightly skewed
* If the absolute value of skew is greater than 1 then very skewed.

```python
descriptiveDF[(descriptiveDF['kurtosis']>1)| (descriptiveDF['kurtosis']<-1) ]
```

```python
#creating a column that generates the entropy of the numeric categorical features

def entNumeric(rows):
  ent={}
  for row in rows.columns:
    ent[row]=stats.entropy(rows[row])

    
    entPD=pd.DataFrame(ent,index=['Entropy'])
  return entPD

```
